```python
import pandas as pd
import sklearn
from sklearn.preprocessing import LabelEncoder

from sklearn import preprocessing

from sklearn.ensemble import BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# DF example:
# df = pandas.read_csv('C:\\Users\\travis\\OneDrive - University of New Haven\\Class Materials\\Sprint 2021\\Week 9\\US Census Data\\census_data.csv')
```

### 1. Return to the US Census Income Dataset from previous weeks
    * We are very familiar with this data at this point.
    * Instantiate the DataFrame in Pandas

```python
df = pd.read_csv('census_data.csv')
df
```

```python
#Dropping nas
df = df.dropna()
df
```

### 2. Split into training/testing sets

```python
# Change of datatype
for column in df.columns:
    if df[column].dtype==type(object):
        le = sklearn.preprocessing.LabelEncoder()
        df[column]=le.fit_transform(df[column])
```

```python
x = df.drop(columns=['income'])
y = df['income']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=555, stratify=y)
```

### 3. Build a DecisionTreeClassifier with 5 fold cross validation   
    * Note results 

```python
dt = DecisionTreeClassifier()
```

```python
results = sklearn.model_selection.cross_val_score(dt, x_train,y_train, cv=5)
results
```

```python
results.mean(), results.std()
```

### 4. Using the bagging technique, build an ensemble with 100 base DecisionTreeClassifier trees
    * Note and compare results 

```python
# BAGGING
# Create a bag of estimators of size 100
dt_bag = BaggingClassifier(base_estimator=dt, n_estimators = 100, random_state=555, n_jobs=-1)

# Fit / Train model
dt_bag.fit(x_train,y_train)

#Results
results = dt_bag.score(x_test, y_test)
results
```

```python
'The results under bagging is slightly higher than the mean of the results under a decision tree classifier.'
```

### 5. Lastly, build a RandomForestClassifier   
    * Note and compare results yet again

```python
rf = RandomForestClassifier()
```

```python
results = sklearn.model_selection.cross_val_score(rf, x_train,y_train, cv=5)
results
```

```python
results.mean(), results.std()
```

```python
'Now that there are the 3 models, we can see the following: The results of random forest classifier is slightly higher than that of bagging (although they are almost similar: only about a 0.0005 difference), and the results of bagging is slightly higher than that of decision tree classifier.'
```
